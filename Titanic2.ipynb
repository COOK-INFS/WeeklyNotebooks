### What do we want to know?

#### Who survived the Titanic sinking?

```python
# We know we need the 'Survived' and the 'Name' columns

who_survived = df.groupby('Survived')['Name'].apply(list)

#print("List of passengers that survived: ", who_survived[1])

# A different method to format the list better
print("Passengers who survived: ")
for name in who_survived[1]:
    print(name)
    
# List of passengers that did NOT survive
print("\nList of passengers that did not survive: ")
for name in who_survived[0]:
    print(name)
```

#### How many Men vs Women survived?

```python
# We know we're going to need the 'Survived' column and the 'Sex' column.

# Group the data by survived and sex, and get the size of each group.
survived_by_sex = df.groupby(['Survived', 'Sex']).size()

print(survived_by_sex)
```
