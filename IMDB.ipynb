# What questions do we want to know?

### What was the name and year of the movie with the highest rating and highest number of votes?

#### Merging data frames

```python
mvRating = pd.merge(dfTitles, dfRatings, on='tconst')
```

```python
mvRating
```

#### Reduce the data to what we need

```python
# Eliminate everything that is NOT a movie.

ratingSort = (mvRating.query('titleType == "movie"'))

ratingSort
```

##### Sort the data in the data frame.

```python
# This shows the sorting by averageRating for all movies.
#ratingSort.sort_values(by='averageRating', ascending=False)

# Sort by multiple columns
ratingSort.sort_values(by=['numVotes', 'averageRating'], ascending=False)
```

#### Create final results

```python
# Create a data frame with the properly queried and sorted data
semi1 = (ratingSort.sort_values(by=['numVotes', 'averageRating'], ascending=False))

# Create a data frame with only the required columns
semi2 = semi1[['primaryTitle', 'startYear', 'averageRating', 'numVotes']]

# Create a data frame with a subset of records
final = (semi2.head(10))

final
```

##### Write the results to a file

```python
final.to_csv('finalResults.csv', index=False)
```
